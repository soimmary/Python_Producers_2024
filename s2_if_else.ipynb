### Логические операторы

У нас может быть сразу несколько условий, которые должны выполняться. В таком случае нам понадобятся логические операторы. Важно помнить, что у операторов есть приоритет выполнения (как в математике сначала идет умножение, а потом складывание).

|Порядок выполнения|Оператор|Значение
|:-----:|:-----:|:-----:|
|1|not| НЕ (отрицание, чаще всего при проверке наличия элемента в другом элементе)|
|2|and| И (обязательное выполнение условий слева и справа от И)|
|3|or| ИЛИ (должно выполниться хотя бы одно из условий слева или справа от ИЛИ)|
