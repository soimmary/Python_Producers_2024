<center><h3>Профессиональная часть (не подготовка к НЭ) – Natural Language Processing </h3> </center>

В частотном анализе выше мы видим, что слова в разных склонениях считаются как разные. Это проблема лемматизации – то есть слова не приведены к исходной форме (так, как они записаны в словаре).

Импортируем библиотеку pymorphy3 – морфологический анализатор для русского языка, написанный на языке Python и использующий словари из [OpenCorpora](https://opencorpora.org/dict.php).

Используем оттуда MorphAnalyzer. Для заданного слова он может найти все возможные формы.

```python
!pip install pymorphy3 # для начала загрузим библиотеку
```

<blockquote>Библиотека — это набор связанных модулей, объединенных вместе для решения определенной задачи или предоставления определенного функционала. Библиотеки можно представить как расширение базового функционала Python, которое помогает разработчикам решать различные задачи без необходимости писать все с нуля.</blockquote>

```python
import pymorphy3 # импортируем библиотеку в нашу тетрадку
morph = pymorphy3.MorphAnalyzer() # создаем морфологический анализатор
```

```python
morph.parse('жди') # используем метод morph.parse('слово'), чтобы найти исходную форму
```

Разберем, что мы получили:

`word = 'жди'` – это введенное нами слова для анализа.

`tag=OpencorporaTag('VERB,impf,tran sing,impr,excl')` – это набор граммем, характеризующих данное слово.

`VERB` – глагол \
`impf` – несовершенный вид \
`tran sing` – переходный глагол в единственном числе \
`impr` – повелительное наклонение \
`excl` – говорящий не включён в действие (жди, ждите)

[Полный список граммем](https://opencorpora.org/dict.php?act=gram)

`normal_form='ждать'` – это каноническая форма слова (например, форма единственного числа, именительного падежа для существительных).

`score=1.0` – это оценка вероятности того, что данный разбор правильный.

`methods_stack=((DictionaryAnalyzer(), 'жди', 973, 11)` – техническая информация о том, как он нашел слово

```python
morph.parse('жди')[0].normal_form # [0] – заходим внутрь списка, забираем нормальную форму
```

Осталось только пропустить наше стихотворение через анализатор.

```python
poetry_dict = {}

for word in poetry_lst:
    word = morph.parse(word)[0].normal_form
    if word not in poetry_dict:
        poetry_dict[word] = 1
    else:
        poetry_dict[word] += 1
```

```python
print(poetry_dict)
```

Согласитесь, что теперь частотный анализ выглядит гораздо лучше! А наш изначальный код изменился всего на одну строку!

И еще один небольшой лайфхак – как быстро делать частотный анализ текста: библиотека `collections`.

```python
from collections import Counter # импортируем "частотный анализатор"
```

```python
for i in range(len(poetry_lst)):
    poetry_lst[i] = morph.parse(poetry_lst[i])[0].normal_form # изменяем наш список
```

```python
print(poetry_lst)
```

```python
Counter(poetry_lst).most_common(20) # печатаем
```
