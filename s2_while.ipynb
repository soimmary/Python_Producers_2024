### Применяйте правильно логику

Распространенной ошибкой при проверке нескольких условий является неправильное применение логики.

Этот пример пытается проверить , если две переменные имеют больше 2. Заявление оценивается как - if (a) and (b > 2).Это дает неожиданный результат , так как bool(a) вычисляется как True , когда не равен нулю. a

```
>>> a = 1
>>> b = 6
>>> if a and b > 2:
...     print('yes')
... else:
...     print('no')

yes
```

Каждую переменную нужно сравнивать отдельно.

```
>>> if a > 2 and b > 2:
...     print('yes')
... else:
...     print('no')

no
```

Другая, аналогичная ошибка допущена при проверке, является ли переменная одним из нескольких значений. Утверждение в этом примере оценивается как - if (a == 3) or (4) or (6).Это дает неожиданный результат , так как bool(4) и bool(6) каждый вычисляться значение True

```
>>> a = 1
>>> if a == 3 or 4 or 6:
...     print('yes')
... else:
...     print('no')

yes
```

Опять каждое сравнение должно быть сделано отдельно

```
>>> if a == 3 or a == 4 or a == 6:
...     print('yes')
... else:
...     print('no')

no
```

# Синтаксис цикла while и его применение

Пусть у нас есть небольшое производство восковых свечей. Однако мы понимаем, что производить свечи можно только до тех пор, пока не кончится воск. Для таких случаев и существует **цикл while**.

    Слово while в переводе с английского означает «пока», поэтому, как правило, этот вид цикла используют в задачах с граничными условиями: повторять цикл необходимо, пока что-то не кончилось, пока что-то не произошло, и тому подобное.
    
Чтобы рассчитать количество свечей, которые можно отлить из купленного воска, необходимо разобраться, как работать с циклом **while**.


#### Рассмотрим конструкцию цикла while.

СИНТАКСИС

````python
while <логическое выражение>:
    <тело цикла>
````

Разберем синтаксис подробно:

|ЭЛЕМЕНТ|ОПИСАНИЕ|
|:---:|:---:|
|<font size="3">**while**</font>|<font size="3">Зарезервированное ключевое слово, с которого начинается цикл|
|<font size="3">**<логическое выражение>**	| <font size="3">Выражение, результатом которого является булево значение `True` или `False`. <br>Если результат <логического выражения> `True`, то цикл продолжает работу. <br>Как только результат <логического выражения> поменялся на `False`, цикл останавливается|
|<font size="3">**:**|<font size="3">Зарезервированный ключевой символ. Обязательно после указания <логического выражения> использовать двоеточие|
|<font size="3">**<тело цикла>**|<font size="3">Инструкции, которые выполняются при каждой итерации. Записывается с отступом относительно ключевого слова while|

А теперь попробуем решить нашу задачу. Для этого мы на каждом этапе цикла будем увеличивать количество изготовленных свечей на 1, а массу воска уменьшать на вес одной свечи, до тех пор пока масса оставшегося воска не станет меньше массы свечи, то есть оставшегося воска не хватит для производства одной свечи.

````python
wax_gr = 5000 #начальное кол-во грамм купленного воска

candle_wax_gr = 120 #грамм воска необходимо для изготовления свечи

candle_cnt = 0 #введем переменную количества свечей (первоначально равна 0)

while wax_gr > candle_wax_gr:

    candle_cnt += 1

    wax_gr -= candle_wax_gr

    print('Уже сделано: ', candle_cnt,' свечей,', 'осталось ',wax_gr, ' грамм воска')
````

Рассмотрим работу написанного нами цикла на каждой итерации:

|. . . . . . . .|	. . . . . . . . . . . .| 	. . . . . . . . . . . . . . . . .|	. . . . . . . . . . . . . . . . .|	. . . . . . . . . . . . .|
|37|	37|	560|	True|	Да|

|39|	39|	320|	True|	Да|

Как только цикл дошел до массы воска в 80 грамм, он остановился, поскольку дальше логическое утверждение принимает значение False. Казалось бы, зачем мы все это выдумывали, если можно было 5 кг разделить на 120 г и получить 41.67, то есть 41 целую свечку и еще немножко воска.

А что, если процесс гораздо сложнее? Например, изготавливаются свечи разного диаметра, требующие разное количество воска, и их необходимо чередовать при производстве, легко ли тогда ответить на вопрос, сколько свечей каждого вида можно произвести? Поэтому для работы автоматизированных задач зачастую используют цикл while или его аналоги в других языках программирования.

    При подсчете количества изготовленных свечей мы на каждой итерации прибавляли к переменной candle_cnt единицу. Такой подход часто используют в работе с любыми циклами, а переменная, которую формируют таким образом (в нашем случае candle_cnt), называется счетчиком.

# Применение цикла while в практических задачах

На практике цикл while, как правило, решает круг довольно узких задач. Давайте рассмотрим их подробнее.

1. Загрузка данных с ограничениями

Многие из вас наверняка работали с Microsoft Excel. Эта программа имеет ограничение на количество строк, хранящихся на одном листе, — примерно 1 млн записей. Такое ограничение не заметно, если вы работаете с небольшими объемами информации, однако становится критичным, если массивы данных измеряются гигабайтами. Поэтому при заливке данных в Excel разработчики часто используют цикл while: пока количество записей на листе меньше 1 млн, данные загружаются, а в случае превышения лимита загрузка данных будет продолжаться на следующем листе.

Код
```
rows = 0

while rows < 1000000:

   save_data  # сохраняем данные на лист

   rows = rows + count(data)  # обновляем количество уже сохраненных записей
```

Загрузка данных в Microsoft Excel — это один из множества примеров задачи загрузки данных куда-либо. Ограничение может возникнуть в количестве записей, файлов, секунд,  объеме информации и так далее. Подавляющее большинство этих задач решаются с помощью цикла while.


2. Технологические процессы с ограничениями

При разборе задач на синтаксис while мы рассматривали процесс изготовления свечей, и для упрощения базового синтаксиса задача сводилась к одному ограничению — количеству воска, однако на практике таких переменных в разы больше, а на больших производствах их количество и вовсе может достигать нескольких сотен. Очевидно, что в таких условиях польза от вычисления искомых значений, а также контроль изготовления продукции на предприятиях с помощью компьютерных программ неоценимы, поэтому программисты часто используют этот тип цикла в практических задачах, связанных с производством.

3. Программы мониторинга

Программы мониторинга служат примером программ с использованием бесконечных циклов. В настоящее время существует множество систем, которые работают без контроля человека. Некоторые из них выполняют свои задачи строго по расписанию — их настроили на работу в определенное время и они ее в это время выполнили. Однако существуют и другие программы — они в случайный момент времени «оживают», проверяют, что вокруг них происходит, выполняют свою работу и снова «засыпают». Такие системы называются системами мониторинга.

Ваша Алиса не работает, пока вы не назвали ее по имени. Все остальное время она "спит".

Как же такие системы узнают, что для них появилась работа? На самом деле никак, это и есть пример бесконечного цикла. Система работает все время: если для нее есть работа, она ее выполняет, если нет, то система проверяет, не появилась ли работа, и так бесконечно.

```
a = 5

while a > 4:

    if find_new_job: #если найдено новое задание

        make_new_job #делай это задание

    else:

        pass
```
        
find_new_job и make_new_job — это «говорящие» имена функций, ранее мы их нигде не указывали.

Как правило, нет необходимости проверять наличие новых заданий каждую секунду. В этом случае для экономии ресурсов программу искусственно «замораживают» с помощью специальных функций заморозки. В Python зачастую используют функцию sleep библиотеки time.
